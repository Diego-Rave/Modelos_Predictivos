### 1. Cargar los datos
A través de la librería **pandas** podemos realizar la carga de datos desde diferentes fuentes de información, en este caso se realizará la carga de un archivo Excel.

```python
# Primero debemos importar la librería pandas
import pandas as pd
# A continuación, se cargan los datos
data=pd.read_excel('./datasets/venta_inmuebles/venta_inmuebles_data.xlsx')
data_dic=pd.read_excel('./datasets/venta_inmuebles/venta_inmuebles_diccionario.xlsx')
```

### 3. Visualizar los datos



```python
# Para esto importamos la libreria a matplotlib
import matplotlib.pyplot as plt
```

```python
# Otro gráfico que podemos generar es uno de dispersión, que nos permitirá ver la relación entre dos variables.
plt.figure(figsize=(10,10))
plt.plot(data['numero_baños'],data['precio'],'.', color='blue')
plt.title('Número de baños vs Precio del Inmueble')
plt.xlabel('Número de baños')  
plt.ylabel('Precio del Inmueble')  
plt.show()
```

```python
# Otra librería que podemos utilizar para graficar es seaborn.  
import seaborn as sns
plt.figure(figsize=(16,6))
plt.tight_layout()
sns.histplot(data['precio'], kde=True, stat='density')
```
